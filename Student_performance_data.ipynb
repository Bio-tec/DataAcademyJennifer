```python
import numpy as np
import altair as alt
import pandas as pd

```

```python
data = pd.read_csv("math.csv")
data.shape
```

```python
data.head(5)
```

```python
data.columns
```

```python
alt.Chart(data).mark_point().encode(
    x='G3',
    y='G1',
    color='sex',
)
```

```python
alt.Chart(data).mark_tick().encode(
    x='G3',
    y='activities:N',
    color='sex:N'
    
)
```

```python
brush=alt.selection(type='interval')
base=alt.Chart(data).add_selection(brush)
points=base.mark_point
```

```python
alt.Chart(data).mark_circle(size=60).encode(
    x='G3',
    y='count(G3)',
    color='sex',
    tooltip=['absences', 'sex', 'G3', 'absences']
).interactive()
```

```python
alt.Chart(data).mark_point().encode(
    x='G3',
    y='absences',
    color='sex',
)
```

```python
alt.Chart(data).mark_point().encode(
    x='G3',
    y='absences',
    color='sex',
).transform_filter(alt.datum.G3==15)
```

```python
alt.Chart(data).mark_bar().encode(
    x= 'address',
    y='count(G3)',
    color='sex:N'

).transform_filter(alt.datum.G3==10)
```

```python

alt.Chart(data).mark_bar().encode(
    alt.X("absences:Q", bin=True),
    y='G3',
    color='sex'
)
```

```python
alt.Chart(data).mark_bar().encode(
    x="G3",
    y="count(G3):Q",
    color="sex:N",
    column="sex"
)
```

```python
alt.Chart(data).mark_bar().encode(
    x=alt.X('G3', bin=alt.Bin(maxbins=20)),
    y='count()',
    color='sex'
    )
```

```python
alt.Chart(data).mark_line().encode(
    alt.X('G2'),
    alt.Y('count(G2)'),
    color='sex:N'
)
```
